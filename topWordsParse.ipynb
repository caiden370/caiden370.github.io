```python
import pandas as pd
df = pd.read_csv('Adsonant-1000-most-common-spanish-words.csv')
```

```python
print(df.columns)
```

```python
INDEX_RANGE = 10000
import json
import os
def create_json(i, word_pairs):
    json_obj = {str(i+INDEX_RANGE):
    {
        "words": word_pairs
    }}
    return json_obj

def write_json(index, json_obj):
    with open(f'spanish-app/src/json-files/topWords/{index + INDEX_RANGE}-topWords.json', 'w') as f:
        s = json.dumps(json_obj).replace('},', '},\n\t\t\t')
        f.write(s)
```

```python
data = df[['Word', 'Translation', 'Rank']].sort_values('Rank', ascending=True).to_numpy()

word_pairs = []
for i in range(data.shape[0]):
    # print(data[i][2])
    word_pairs.append({"spanish": f"{data[i][0]}", "english": f"{data[i][1]}"})
    if i % 50 == 49:
        print(i//50)
        write_json(i//50, create_json(i//50, word_pairs))
        word_pairs = [];
    
```
